## sentiment analysis pipeline

```python
!pip -qq install transformers
```

```python
from transformers import pipeline
import pandas as pd
from pprint import pprint
from sklearn.metrics import classification_report
```

```python
classifier = pipeline("sentiment-analysis", device=0)
```

```python
!wget -nc https://lazyprogrammer.me/course_files/AirlineTweets.csv
```

```python
df = pd.read_csv('AirlineTweets.csv')
df
```

```python
df.airline_sentiment.value_counts()
```

```python
df.text
```

```python
df2 = df[['airline_sentiment','text']]
```

```python
df3 = df2[df2['airline_sentiment']!='neutral']
df3
```

```python
df3.text.to_list()
```

```python
preds = classifier(df3.text.to_list())
preds
```

```python
len(preds)
```

```python
preds[0]['label']
```

```python
y_pred =  [1 if pred['label']=='POSITIVE' else 0 for pred in preds]
y_pred[:5]
```

```python
df3.airline_sentiment.map({'positive':1, 'negative':0})
```

```python
y_true = df3.airline_sentiment.map({'positive':1, 'negative':0}).tolist()
y_true[:5]
```

```python
print(classification_report( y_true , y_pred))
```

## zero-shot pipeline

```python
zshotclf = pipeline("zero-shot-classification", device=0)
```

```python
!wget -nc https://lazyprogrammer.me/course_files/AirlineTweets.csv
```

```python
df = pd.read_csv('AirlineTweets.csv')
df
```

```python
df.airline_sentiment.value_counts()
```

```python
df.text
```

```python
df2 = df[['airline_sentiment','text']]
```

```python
df3 = df2[df2['airline_sentiment']!='neutral']
df3
```

```python
reviews = df3.text.to_list()
reviews[:5]
```

```python
labels = df3.airline_sentiment.unique()
labels
```

```python
preds = zshotclf(reviews, candidate_labels= labels)
preds
```

```python
preds[0]['labels'][0]
```

```python
y_pred =  [1 if pred['labels'][0]=='positive' else 0 for pred in preds]
y_pred[:10]
```

```python
y_true = df3.airline_sentiment.map({'positive':1, 'negative':0}).tolist()
y_true[:10]
```

```python
print(classification_report( y_true , y_pred))
```
